```python
import requests
import bs4
import urllib.parse
import time
import random
import re
from furl import furl
```

# Define Variables

```python
# Define keywords
keyword_primary = "iOS"
keyword_secondary_list = ['遠端', '遠距', '在家']
```

```python
url_original = 'https://www.1111.com.tw/search/job?ks=ios&page=1&ps=100'
```

# Gather web page results

```python
f = furl(url_original)
```

```python
f.args['ks'] = keyword_primary 
```

```python
page_soups = []
page_count = 1
total_page_count = None
while True:
    print(f'Start fetching page {page_count}...')
    f.args['page'] = page_count
    print(f.url)
    
    r = requests.get(f.url) 
    if r.status_code != 200:
        break
    
    soup = bs4.BeautifulSoup(r.text, 'lxml')
    page_soups.append(soup)

    # find total pages count if not found yet
    if total_page_count is None:
        total_page_count = len(soup.find_all('select', class_='custom-select srh-footer__content-page')[0].find_all('option'))
    if total_page_count == page_count:
        break
    
    time.sleep(random.randint(5, 10))
    page_count += 1
```

```python
print(f'Found {len(page_soups)} Pages.')
```

# Use secondary keywords to filter the results

```python
regex = '|'.join(keyword_secondary_list)
flags = re.IGNORECASE
```

```python
job_urls = []
for soup in page_soups:
    a_list = soup.find_all('a', class_='item__job-info--link item__job-position0--link')
    job_urls.extend([urllib.parse.urljoin(url_original, a['href']) for a in a_list])
print(f'Totally {len(job_urls)} jobs to filter.')
```

```python
results = []
count = 0
for url in job_urls:
    r = requests.get(url)
    print(f'Start fetching url: {url}...')

    if re.search(regex, r.text.replace('ios-app', 'app'), flags = re.IGNORECASE):
        results.append(url)

    count += 1
    print(f'Filter progress: {round(count/len(job_urls)*100, 2)}%')

    time.sleep(random.randint(5, 10))
```

```python
print(f'Found {len(results)} results with the primary keyword "{keyword_primary}" and the secondary keywords "{keyword_secondary_list}".')
```

# Show the results

```python
for result in results:
    print(result)
```
